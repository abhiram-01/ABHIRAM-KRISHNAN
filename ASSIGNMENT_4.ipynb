##Question 2

```
import numpy as np
from sklearn.linear_model import LinearRegression


X= np.array(students.iloc[:,5].values)
y = np.array(students.iloc[:,6].values)

X= X.reshape(-1, 1)
y= y.reshape(-1, 1)

print( 'X' , X [:5] )
print( 'y' , y [:5] )
```

##Question 3

```
model = LinearRegression()
model.fit( X , y )
model.score( X , y )
```
